```python
import pandas as pd
import numpy as np
import plotly.express as px
import folium
import geopandas as gpd
```

```python
# importando arquivos finais

rota_organizada = pd.read_csv('/content/dados_finais_ordenados.csv')
planilha_supref = pd.read_excel('/content/Sem título 1.xls')
```

```python
planilha_supref.to_csv('Pedidos por subprefeitura.csv')
```

```python
# SEstabelecendo 'subprefeitura' como a coluna comum entre os dois DataFrames
df_cruzado = pd.merge(rota_organizada, planilha_supref, on='subprefeitura')

df_cruzado = df_cruzado.rename(columns={'quantidade_x': 'quantidade_escola', 'quantidade_y': 'quantidade_subprefeitura' })

# Calcula o número de escolas únicas por subprefeitura
numero_escolas2 = df_cruzado.groupby('subprefeitura')['escolas_postos'].nunique()

# Transforma a série em um DataFrame
numero_escolas2_df = numero_escolas2.reset_index()

# Mescla os dois daframes
df_cruzado = pd.merge(df_cruzado, numero_escolas2_df, on='subprefeitura')

#media de entregas por sunprefeitura
df_cruzado['media_escolas'] = df_cruzado['quantidade_subprefeitura'] / df_cruzado['escolas_postos_y']

# Cria um DataFrame com o número de escolas e a média de volume entregue para cada subprefeitura
info_subprefeitura = df_cruzado.groupby('subprefeitura')[['escolas_postos_y', 'media_escolas']].mean().reset_index()

df_cruzado
```

```python
# Gerando gráficos-exemplo

fig = px.bar(planilha_supref, x='subprefeitura', y='quantidade', title='Quantidade de pedidos por Subprefeitura')

# Adiciona dicas (tooltips) com o número de escolas e a média de volume entregue
fig.update_traces(
    hovertemplate=(
        "Subprefeitura: %{x}<br>"
        "Quantidade: %{y}<br>"
        "Número de Escolas: %{customdata[0]}<br>"
        "Média de Volume Entregue: %{customdata[1]:.2f}"
    ),
    customdata=info_subprefeitura[['escolas_postos_y', 'media_escolas']]
)

fig.show()

```
